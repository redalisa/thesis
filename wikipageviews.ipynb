```python
species = pd.read_csv("csv/speciesinfo.csv", sep = ";")
```

```python
name = "Panthera tigris"

address = "https://wikimedia.org/api/rest_v1/metrics/pageviews/top-by-country/en.wikipedia.org/wiki/Tiger/all-access/2016"

headers = CaseInsensitiveDict()
headers["Accept"] = "application/json"
# Personal username for identification for the Wikipedia API
headers = {'User-Agent': 'reddalisa/3.10.6 (alisa.redding@helsinki.fi)'}
    
resp = requests.get(address, headers=headers)
details = resp.json()
print(details)
```

```python
#ONE_MINUTE = 60
#MAX_CALLS_PER_MINUTE = 90

#@sleep_and_retry
#@limits(calls=MAX_CALLS_PER_MINUTE, period=ONE_MINUTE)
# https://akshayranganath.github.io/Rate-Limiting-With-Python/
def WikiPageView(name):
    
    # Calling monthly page views of each species 
    address = "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia.org/all-access/user/" + name + "/monthly/2015010100/2020123100"

    headers = CaseInsensitiveDict()
    headers["Accept"] = "application/json"
    # Personal username for identification for the Wikipedia API
    headers = {'User-Agent': 'reddalisa/3.10.6 (alisa.redding@helsinki.fi)'}
    
    resp = requests.get(address, headers=headers)
    details = resp.json()
    
    
    return details 
      
    
# Create an empty list to store the monthly pageviews into
newframe = []  

# Loop over the pd series of species names and call the wikipage function
for name in species['scientific_name']:
    # Spaces are replaced with an underscore for the wikipedia API
    name = name.replace(" ", "_")
    result = WikiPageView(name)
    
    try:
        for n in range(len(result['items'])):
            name = result['items'][n]['article']
            date = result['items'][n]['timestamp']
            # Remove 00 at marking hour
            date = date[:-2]
            view = result['items'][n]['views']
    
            data = {'name': name, 'date': date, 'views': view}
            newframe.append(data)
    except KeyError: 
        pass
```

```python
len(newframe)
```

```python
# 
wikipage = pd.DataFrame(newframe)
wikipage["datetime"] = wikipage["date"].apply(pd.to_datetime)
wikipage['scientific_name'] = wikipage["name"].apply(lambda x: x.replace("_", " "))
```

```python
# 
wikipage.sort_values(by="views")
```

```python
wikipage.to_csv("csv/wikipageviews_species.csv", sep = ";", index = False)
```

```python
top20wiki = wiki.groupby('scientific_name')['views'].sum().nlargest(20).reset_index()
```

```python
# Read in the CITES data to join with the top 10 (or 20) most reported on species 
cites1 = pd.read_csv("csv/cites1.csv", delimiter = ';')

# Keep only most relevant columns
cites1 = cites1[['Kingdom', 'Class', 'Order', 'Family', 'Genus', 'Species', 'Scientific Name', 'Listing', 'NativeDistributionFullNames', 'Extinct_Distribution']]

cites1 = cites1.rename(columns = {'Scientific Name' : 'scientific_name'})
```

```python
top20wiki
```

```python
# Do the same for the entire dataframe
top20wiki = top20wiki.merge(cites1, on = 'scientific_name', how = 'left').drop_duplicates().reset_index(drop=True)
```

```python
top20wiki
```

```python
import seaborn as sns

ax = sns.barplot(
    x = 'scientific_name',
    y = 'views',
    hue = 'Kingdom',
    data = top20wiki, 
    width = 0.7, palette = 'Set3', dodge = False
)
ax.set_xlabel('Scientific Name', fontsize=9)
ax.set_ylabel('WikiPage Views', fontsize=9)


# Add suptitle above the title
plt.suptitle("Top 20 Most Viewed Species of CITES Appendix 1 on Wikipedia\n", fontsize=12)

# Add title
#ax.title("Realized by the Python Graph Gallery", color="grey", style='italic')
ax.set_title('Page views 2015-2020, English Wiki', fontsize = 10, color="grey", style='italic')


ax.tick_params(axis='x', rotation=85, length = 9, labelsize = 7, left = True)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.savefig('figures/top_20_species_wiki.png', bbox_inches = 'tight')
```
