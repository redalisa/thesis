```python
species = pd.read_csv("csv/fullspeciesinfo.csv", sep = ";")
species
```

```python
from requests.structures import CaseInsensitiveDict
from ratelimit import limits, RateLimitException, sleep_and_retry

ONE_MINUTE = 60
MAX_CALLS_PER_MINUTE = 95

@sleep_and_retry
@limits(calls=MAX_CALLS_PER_MINUTE, period=ONE_MINUTE)
# https://akshayranganath.github.io/Rate-Limiting-With-Python/
def pyGoogleSearch(name):
    
    # Personal API to call the google search results for each species name 
    # Only request number of hits with the term charismatic and quotations around the scientific name
    address = "https://customsearch.googleapis.com/customsearch/v1?cx=c687316f685e24c21&q=%22" + name + "%22&fields=queries(request(totalResults))&key=AIzaSyArW4JPnZFaX7DIUgtBlmclSaQBrEppPFo"

    headers = CaseInsensitiveDict()
    headers["Accept"] = "application/json"
    
    resp = requests.get(address, headers=headers)
    details = resp.json()
    #print(details)
    
    for x in details['queries']['request'][0].values():
        return x

# Define a pandas series with unique instance of species in the dataset
#species = pd.Series(species['scientific_name'].unique())
# Create an empty list to store the google search results in
hits = []  

# Loop over the pd series of species names and call the google api function
for name in species['scientific_name']:
    result = pyGoogleSearch(name)
    hits.append((result, name))
```

```python
# create DataFrame using data
google_hits = pd.DataFrame(hits, columns =['hits', 'scientific_name'])

google_hits["hits"] = google_hits["hits"].apply(pd.to_numeric)

google_hits = google_hits.sort_values(by = ["hits"], ascending=False).reset_index(drop = True)
```

```python
import numpy as np
from sklearn.cluster import MeanShift, estimate_bandwidth

# #############################################################################
X = np.reshape(list(google_hits['hits']), (-1, 1))
# #############################################################################
# Compute clustering with MeanShift

# The following bandwidth can be automatically detected using
# bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=100)

ms = MeanShift(bandwidth=None, bin_seeding=True)
ms.fit(X)
labels = ms.labels_
cluster_centers = ms.cluster_centers_

labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("number of estimated clusters : %d" % n_clusters_)
print(labels)
```

```python
google_hits['Charisma'] = 0

for x in google_hits["gl_hits"]:
    if x > google_hits["gl_hits"].mean():
        google_hits.loc[google_hits['gl_hits'] == x, 'Charisma'] = 1
    
```

```python
#df['COUNTER'] =1       #initially, set that counter to 1.
grouped = google_hits.groupby(['Charisma']) 
# Calculate the totals of the group species
charisma = grouped.scientific_name.nunique()


ax = charisma.plot.bar(figsize = (5,3), fontsize = 5, color = ['lightseagreen', 'wheat'])
ax.set_xlabel('Charisma', fontsize=7)
ax.set_ylabel('Species Count', fontsize=7)
ax.set_title('Charismatic and Non-Charismatic Species in Dataset', fontsize = 9)
ax.tick_params(axis='x', rotation=0, left = True, pad = 2)
```
